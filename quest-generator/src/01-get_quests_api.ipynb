```python
import requests
import json
import time
```

---

# Skyrim

```python
skyrim_quests = json.loads(requests.get("https://elderscrolls.fandom.com/api/v1/Articles/List?category=Skyrim%3A_Quests&limit=500").text)
```

```python
quest_ids = [item["id"] for item in skyrim_quests['items']]
```

```python
all_skyrim_quests = {}

for i, quest_id in enumerate(quest_ids[1:]):
    try:
        print(f"{i}/{len(quest_ids)-1}")
        
        quest_article = json.loads(requests.get(f"https://elderscrolls.fandom.com/api/v1/Articles/AsSimpleJson?id={quest_id}").text)
        all_skyrim_quests[quest_id] = quest_article["sections"]
        
        time.sleep(5)
    except Exception as e:
        print(f"Failed to download quest {quest_id}: {e}")
```

```python
with open("skyrim_quests.json", "w") as outfile:
    json.dump(all_skyrim_quests, outfile)
```
